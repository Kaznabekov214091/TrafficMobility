```python
from itertools import product
from datetime import date
from datetime import datetime
import pandas as pd
from pyspark.sql.types import StructType, StructField, StringType, IntegerType, BooleanType, TimestampType
```

```python
taxi_types = ["yellow", "green"]
start_dates = {
    "yellow": datetime(2009, 1, 1),
    "green": datetime(2014, 1, 1),
}

end_date = datetime(datetime.today().year, 10, 1)
rows = []

for taxi in taxi_types:
    months = pd.date_range(start=start_dates[taxi], end=end_date, freq="MS")
    for m in months:
        ingested_at_placeholder = (m - pd.DateOffset(months=1)).to_pydatetime()
        rows.append((
            taxi,
            m.year,
            f"{m.month:02d}",   
            False
        ))

schema = StructType([
    StructField("taxi_type", StringType(), True),
    StructField("year", IntegerType(), True),
    StructField("month", StringType(), True),
    StructField("ingested", BooleanType(), True),
])
df = spark.createDataFrame(rows, schema)
df.write.format("delta").mode("overwrite").saveAsTable("bronze.dbo.taxi_config")
```
